```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/tháng.csv')


plt.figure(figsize=(25, 10))
plt.plot(df['month'], df['total_payment'], marker='o', label='Doanh thu')
plt.title('Doanh thu theo tháng')
plt.xlabel('Tháng')
plt.ylabel('Doanh thu')
plt.grid(True)

for i, txt in enumerate(df['total_payment']):
    plt.text(df['month'][i], txt + 150, f'{txt}', ha='center', va='bottom')

plt.legend()
plt.show()
```
